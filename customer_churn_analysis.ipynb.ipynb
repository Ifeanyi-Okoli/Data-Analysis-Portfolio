```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('/content/Telco-Customer-Churn.csv')

# Data Cleaning
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df['TotalCharges'].fillna(df['TotalCharges'].mean(), inplace=True)
df = df.drop(columns=['customerID'])

# Exploratory Data Analysis
plt.figure(figsize=(6, 4))
sns.countplot(x='Churn', data=df)
plt.title('Churn Distribution')
plt.xlabel('Churn')
plt.ylabel('Count')
plt.savefig('churn_distribution.png')
plt.show()

plt.figure(figsize=(8, 6))
sns.boxplot(x='Churn', y='tenure', data=df)
plt.title('Tenure vs. Churn')
plt.xlabel('Churn')
plt.ylabel('Tenure (Months)')
plt.savefig('tenure_vs_churn.png')
plt.show()

numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = df[numeric_cols].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numeric Features')
plt.savefig('correlation_matrix.png')
plt.show()

print("Summary of Findings:")
print("- Customers who churn tend to have shorter tenure (median ~10 months) compared to non-churning customers (~40 months).")
print("- MonthlyCharges and TotalCharges show moderate correlation, indicating higher charges may relate to longer tenure.")
print("- The dataset has ~26% churn rate, suggesting a significant portion of customers leave the service.")

df.to_csv('cleaned_telco_churn.csv', index=False)
```
